By using `if` function and rules stated above, determine who is going to win the game. It will be either the computer or the player.

```python
#TODO

def determine_winner(player_chose, computer):
    if player_chose == computer:
        return "Tie"

    elif player_chose == "burung":
        if computer == "batu":
            return "Computer wins"
        else:
            return "Player wins"

    elif player_chose == "batu":
        if computer == "air":
            return "Computer wins"
        else:
            return "Player wins"

    elif player_chose == "air":
        if computer == "burung":
            return "Computer wins"
        else:
            return "Player wins"

    else:
        return "Invalid choice"

print(player_chose, " vs ", computer)
print(determine_winner(player_chose, computer))
```
